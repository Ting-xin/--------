```python
import json

with open('./长三角科研机构知网论文数.json', 'r', encoding='utf-8') as file:
  school_list = json.load(file)

school_list
```

```python
import csv

result = {}
for school in school_list:
  if '中文文献数' not in school:
    result[school['市']] = 0
    continue
  if school['市'] in result:
    result[school['市']] += school['中文文献数']
  else:
    result[school['市']] = school['中文文献数']


# 创建 CSV 文件并写入数据
with open('长三角各城市科研机构论文树.csv', 'w', encoding='gbk', newline='') as f:
    writer = csv.writer(f)

    # 写入 CSV 文件的标题行
    writer.writerow(['市', '知网中文文献数'])

    # 遍历字典的列表，并将每个字典的数据写入 CSV 文件中
    for item in result.keys():
        row = [item, result[item]]
        writer.writerow(row)
```
